```python
from quadcubicViz.secondordercurves import Ellipse

ellipse = Ellipse(5, 3, x0=2, y0=3)
ellipse.describe(3, 4)
ellipse.plot()

```

```python
from quadcubicViz.thirdordercurves import CubicParabola

cubic_parabola = CubicParabola(A=1, B=-2, C=3, D=4)
cubic_parabola.properties()
cubic_parabola.plot()
```

```python
from quadcubicViz.thirdordercurves import FoliumOfDescartes
fod = FoliumOfDescartes(4)
fod.plot()
```

```python
from quadcubicViz.thirdordercurves import CissoidOfDiocles
cod = CissoidOfDiocles(5)
cod.area()
cod.plot()
```

```python
from quadcubicViz.thirdordercurves import ConchoidOfDeSluze

cods = ConchoidOfDeSluze(8)
cods.plot()
parametrs = [-6, -3, -1, 0, 3, 6]
diff_a = ConchoidOfDeSluze(parametrs)
diff_a.plots_with_diff_a(row=2, col=3)

```

```python
from quadcubicViz.secondordercurves import Parabola

# Create a parabola y = x² - 4x + 3
p = Parabola(15, 23, -23)

# Visualize the parabola
p.plot()
```

```python
from quadcubicViz.thirdordercurves import SerpentiveCurve
# Створюємо екземпляр класу
curve = SerpentiveCurve(a=1, b=1)

# Візуалізуємо криву
curve.plot(x_range=(-5,5), num_points=1000,color='red')
```

```python
from quadcubicViz.thirdordercurves import TridentCurve

trident = TridentCurve(a=1, b=1, c=-1, d=1)
trident.plot(x_range=(-3,3),color="green")
```

```python
from quadcubicViz.thirdordercurves import TschirnhausenCubic

# Створюємо екземпляр класу
curve = TschirnhausenCubic(a=1.0)

curve.plot((-3,3))
```
